## Feaature Engineering 1.creating salary bands 2.experience level bins

```python
df['salary_bands'] = pd.cut(
    df['average_salary'],
    bins=[0,50000,100000,150000,300000],
    labels=['low','medium','high','very_high']
)
```

```python
df['experience_level'] = pd.cut(
    df['years_experience'],
    bins=[0,3,7,15,50],
    labels=['entry','mid','senior','expert']
)
df['experience_level'].value_counts()
```

## 1.which jobs are at highest AI automation risk

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
sns.set_theme()
```

```python
risk_counts = df['risk_category'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=risk_counts.index,y=risk_counts.values,color='blue')
plt.title("Distribution of Jobs by AI Automation Risk")
plt.xlabel("Risk Category")
plt.ylabel("Number of Jobs")
plt.show()
```

## 2. Salary vs Automation Risk--Does high salary reduces automation risk


```python
## median salary comparison across risk groups
plt.Figure(figsize=(7,4))
sns.boxplot(
    data=df,x='risk_category',y='average_salary'
)
plt.title("Average Salary Distribution by Automation Risk")
plt.xlabel("Risk Category")
plt.ylabel("Average Salary")
plt.show()
```

## 3.Exp level Vs AI Exposure

```python
## does senior roles face more or less AI EXPOSURE

exp_ai = df.groupby('experience_level')['ai_exposure_index'].mean().reset_index()
plt.Figure(figsize=(7,4))
sns.barplot(data=exp_ai,
            x='experience_level',
            y='ai_exposure_index')
plt.title("Average AI Exposure by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("AI Exposure Index")
plt.show()
```

###### 4.Salry bands vs Risk category

```python
## how does automayion risk change salary bands

salary_risk = pd.crosstab(
    df['salary_bands'],
    df['risk_category'],
    normalize='index'
)

salary_risk.plot(
    kind='bar',
    stacked=True,
    figsize=(9,6)
)

plt.title("Automation Risk Distribution Across Salary Bands")
plt.xlabel("Salary Band")
plt.ylabel("Proportion of Jobs")
plt.legend(title="Risk Category")
plt.show()
```

## 5.skill impact anlysis

```python
## which skill relate with lower automation risk

skill_cols = [col for col in df.columns if col.startswith('skill_')]

skill_risk = df.groupby('risk_category')[skill_cols].mean()

plt.figure(figsize=(10,5))
sns.heatmap(
    skill_risk,
    annot=True,
    cmap='coolwarm'
)
plt.title("Average Skill Scores by Automation Risk Category")
plt.xlabel("Skills")
plt.ylabel("Risk Category")
plt.show()
```

## 6. correlation heatmap

```python
corr = df[
    ['average_salary', 'years_experience',
     'ai_exposure_index', 'automation_probability_2030']
].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Correlation Between Key Numerical Variables")
plt.show()
```

## 7. Does higher education protect jobs from AI auomation

```python
## education level Vs risk category 

edu_risk_pct = pd.crosstab(
    df['education_level'],
    df['risk_category'],
    normalize='index'
)

## rows-edu levels , clmns-risk cate

edu_risk_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(9,6)
)



plt.title("Education Level vs Automation Risk Distribution")
plt.xlabel("Education Level")
plt.ylabel("Percentage of Distribution")
plt.legend(title="Risk Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## 8 .Quadrant Analysis: Salary vs Automation Risk


## Which high-paying jobs are still vulnerable to automation?

```python
## 4-quadrant scatter plot

# --columns used
quad_df = df[['job_title','average_salary','automation_probability_2030']].copy()

# -- medians for quadrant lines --
median_salary=quad_df['average_salary'].median()
median_risk=quad_df['automation_probability_2030'].median()

# -- create scatter plot --
plt.figure(figsize=(9,6))
sns.scatterplot(
    data=quad_df,
    x='average_salary',
    y='automation_probability_2030',
    alpha=0.6
)

# -- adding quadrants / median lines --
plt.axvline( median_salary, color ='red', linestyle = '--', label='Median Salary')
plt.axhline( median_risk, color ='blue', linestyle = '--', label='Median Automation Risk')

# --- Annotate quadrants ---
plt.text(median_salary * 1.05, median_risk * 1.05,
'Vulnerable\n(High Pay / High Risk)', color='darkred')


plt.text(median_salary * 1.05, median_risk * 0.25,
'Safe Havens\n(High Pay / Low Risk)', color='darkgreen')


plt.text(median_salary * 0.25, median_risk * 1.05,
'At-Risk\n(Low Pay / High Risk)', color='brown')


plt.text(median_salary * 0.25, median_risk * 0.25,
'Stable\n(Low Pay / Low Risk)', color='darkblue')

# --final steps--
plt.title('Quadrant Analysis: Salary vs Automation Risk')
plt.xlabel('Average Salary')
plt.ylabel('Automation Probability (2030)')
plt.legend()
plt.tight_layout()
plt.show()
```

## Skill Predictors of Low Risk Jobs

```python
#--identify skill columns
skill_cols = [col for col in df.columns if col.startswith('skill_')]
#--filter only low risk jobs
low_risk_df = df[df['risk_category'] == 'Low']
#--compute avg skill importance
skill_means = low_risk_df[skill_cols].mean().sort_values()

## horizongal bar
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
skill_means.plot(kind='barh', color='steelblue')

plt.title('Skill Predictors of Low Automation Risk Jobs')
plt.xlabel('Average Skill Score (Low-Risk Jobs)')
plt.ylabel('Skill')
plt.show()


```

```python
# -- fetch skill columns
skill_cols = [col for col in df.columns if col.startswith('skill_')]
# -- split dataset by risk category
low_risk_df = df[df['risk_category'] == 'Low']
high_risk_df = df[df['risk_category'] == 'High']
# -- compute mean skill scores
low_risk_means = low_risk_df[skill_cols].mean()
high_risk_means = high_risk_df[skill_cols].mean()
# -- compute skill difference
skill_comparison = (low_risk_means - high_risk_means).sort_values()
# -- convert to dataframe
comparison_df = pd.DataFrame({
    'skill': skill_comparison.index,
    'difference_low_minus_high': skill_comparison.values
})

# -- visualization
plt.figure(figsize=(9, 5))

sns.barplot(
    data=comparison_df,
    x='difference_low_minus_high',
    y='skill',
    palette='coolwarm'
)

plt.axvline(0, color='black', linestyle='--')  # zero reference line

plt.title('Skill Comparison: Low-Risk vs High-Risk Jobs')
plt.xlabel('Average Skill Difference (Low Risk  High Risk)')
plt.ylabel('Skill')
plt.tight_layout()
plt.show()
```
